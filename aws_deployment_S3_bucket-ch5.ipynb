```python
import subprocess

s3_bucket_name = "mlops-sagemaker-runs-rhys"
mlruns_direc = "./mlruns/"

output = subprocess.run(["aws","s3","sync","{}".format(mlruns_direc), "s3://{}".format(s3_bucket_name)],
                       stdout=subprocess.PIPE,encoding='utf-8')

print(output.stdout)
print("\nSaved to bucket: ", s3_bucket_name)
```
